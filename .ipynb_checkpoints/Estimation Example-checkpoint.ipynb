## d-max-decreasing Neural Networks

In this notebook we will consider the problem of estimating a multivariate extreme value (MEV) distribution. Recall that an MEV can be specified by an extreme value copula and has the form of
$$
\mathbb{P}(F_1(X_1) < u_1, \ldots, F_d(X_d) < u_d) := \exp\left(\sum_{j=1\ldots d} u_j A\left(\frac{\log u_1}{\sum_{j=1\ldots d} u_j}, \ldots, \frac{\log u_d}{\sum_{j=1\ldots d} u_j}; \theta \right)\right)
$$
where $A(\cdot; \theta)$ is a Pickands dependence function parameterized by $\theta$.

In this notebook, we will parameterize $A$ using a d-max-decreasing neural network (dMNN) following the paper [Modeling Extremes with d-max-decreasing Neural Networks](https://arxiv.org/abs/2102.09042). 

#### Structure of the dMNN

The main building block of a dMNN corresponds to the spectral representation of $A$ given by
$$
A(s_1, \ldots, s_d) = \mathbb{E}_{w \sim \Lambda} \left [ \max_{j = 1, \ldots, d} s_i w_i \right]
$$
where $\Lambda$ is the spectral measure. 
One can think of this as a single layer, infinite width network with the $\max$ activation function and weights sampled according to $\Lambda$. 
The dMNN is defined in the `nets.py` file as the class `dMNN`.

#### Parameter Estimation
Since differentiating the full CDF is highly prone to numerical errors, we consider a particular property of extreme value distributions. 
Let $M$ be a sample of componentwise maximum and let $\tilde{M}_j = -\log F_j (M_j)$ be the log of the margin transformed maximum. 
Define the random variable $Z_w$ as
$$
Z_w = \bigwedge_{j=1\ldots d} \tilde{M}_j/w_j, \quad w \in \Delta_{d-1}. 
$$
Then $Z_w \sim \text{Exp}(A(w))$. 
This transformation allows us to use the maximum likelihood estimate for a 1-dimensional random variable for the full d-dimesional vector by sampling along the simplex $\Delta_{d-1}$. 

```python
import torch

from pytorch_lightning import Trainer

from datasets import SL, ASL
from utils import init_estimators, rand_simplex
from pickands import ConditionalPickandsModule
from nets import dMNN, ICNN

%matplotlib inline
import matplotlib.pyplot as plt
```

Here we collect our data. This is based on data from the asymmetric logistic distribution with $\alpha = 0.5$. In general, this is the function we want to define for training on a custom dataset.  

```python
from torch.utils.data import DataLoader

# data dimension
d = 50

# number of samples
n_samp = 10

# dependency parameter
alpha = 0.5

dataset = SL(d, n_samp=n_samp, alpha=alpha)
val_dataset = SL(d, n_samp=n_samp, alpha=alpha)

train_loader = DataLoader(dataset, batch_size=1000, shuffle=False)
val_loader   = DataLoader(val_dataset, batch_size=1000, shuffle=False)
```

Add some plotting code for the 2d marginals. 

```python
import torch.nn.functional as F
import numpy as np


import seaborn as sns 

def get_all_bi_extremal_coeffs(a, d, labels=None, a_true=None, d_use=None):
    if d_use == None:
        d_use = d // 2

    plt.style.use('seaborn-darkgrid')
    
    with torch.no_grad():

        c_ind = 0

        for s1 in range(d_use):
            for s2 in range(d_use):
                if s2 < s1:
                    continue

                x  = torch.linspace(0, 1-1e-8, 100).cpu()
                x_ = torch.stack((x, 1 - x), dim=1)
                
                one = torch.zeros(x.shape[0],d).cpu()
                one[:,s1] = x_[:,0]
                one[:,s2] = x_[:,1]

                dep = a(one)

                if s1 < s2:
                    plt.plot(x.detach().cpu(), dep.detach().cpu().numpy(), color='blue', alpha=0.5)
                    c_ind += 1
                    if a_true is not None:
                        plt.plot(x.detach().cpu(), a_true(one).detach().cpu().numpy(), color='black', alpha=0.5)

        ax = plt.gca()
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        plt.xlabel(r'$w \in \Delta_1$', fontsize=15)
        plt.ylabel(r'$A(w, 1-w)$', fontsize=15)
        plt.xticks(fontsize=15)
        plt.yticks(fontsize=15)
        plt.tight_layout()
        plt.show()
```

We will now define some of the main parameters needed for our estimation task. In the original paper we simply used the MLE on the exponential distribution described above. However, this isn't exactly necessary. In fact, any non-parametric estimator can be used as a loss where we simply project the estimator onto the space of d-max-decreasing functions through the dMNN. In the example below we use a linear combination of the MLE and the Pickands estimator (which, theoretically, correspond to the same thing). Using different losses may be beneficial in certain cases, particularly if optimization is difficult. Ultimately, whatever loss is used is guaranteed to recover a function that satisfies the properties of the Pickands dependence function.

```python
%%capture --no-display

# Trainer parameters
trainer_params = {
        'gpus' : 0,
        'max_epochs' : 1000,
        'progress_bar_refresh_rate' : 100,
        }

# Conditional Pickands Module Parameters
model_params = {
        'lr' : 1e-2,
        'l1' : 0,
        'survival' : False,
        'use_swa' : False,
        'loss_type' : 'mle-pick'
        }

# dMNN Parameters
arch_params = {
        'input_size' : d,
        'depth' : 1,  
        'width' : 100,
        'cond_size' : 0,
        'cond_width': 0,
        }

# initialize the network
net = dMNN(**arch_params)

# initialize the training logic
cpm = ConditionalPickandsModule(net, **model_params)

# initialize the trainer
trainer = Trainer(**trainer_params)

# fit the model
trainer.fit(cpm, train_loader)
```

Now that we've trained, we can plot 2d marginal subsets to get an idea if the properties of the Pickands function are enforced. The key aspect of d-max-decreasing results in consistent margins where subsets of margins remain valid Pickands functions.

```python
get_all_bi_extremal_coeffs(cpm.a, d, a_true=dataset.pickands)
```

Now let's check the MSEs between the different methods.

```python
w   = rand_simplex(10000, d)
mse = F.mse_loss(cpm.a(w), dataset.pickands(w).unsqueeze(-1))
print('dMNN MSE : {:.3E}'.format(mse.item()))
```

Note that all the properties of the Pickands dependence function are retained on the margins. Now let's compare to the other nonparametric estimators.

```python
np_list = ['cfg', 'bdv']
np_estimators = init_estimators(np_list,
            dataset.data,
            survival=False,
            est_F=dataset.F)

for idx, estimator in enumerate(np_estimators):
    get_all_bi_extremal_coeffs(estimator, d, a_true=dataset.pickands)
    mse = F.mse_loss(estimator(w), dataset.pickands(w))
    print('{} MSE : {:.3E}'.format(np_list[idx], mse.item()))
    
```

From these plots, we can see that the properties are not enforced, specifically the property of convexity is violated with both the CFG and BDV estimators.
